```python
def get_headlines(data):
    headlines =[]
    
    for i in range(0, len(data.index)):
        
        headlines.append(" ".join(str(x) for x in data.iloc[i,1:]))

        
    return headlines
```

```python
para_data = get_headlines(df)
para_data[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer   # For creating Bag of Words
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv = CountVectorizer(max_features=10000,ngram_range=(1,1))    
#from sklearn.feature_extraction.text import TfidfVectorizer  # For creating Bag of Words
#cv = TfidfVectorizer(max_features=1500)

x = cv.fit_transform(para_data).toarray()
y = df['Label'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 2)
```

```python
from sklearn.naive_bayes import MultinomialNB
model =  MultinomialNB(alpha= 1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
matrix  = confusion_matrix(y_pred,y_test)
```

```python
matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred,y_test)
accuracy
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
linear_clf = PassiveAggressiveClassifier()
```

```python
linear_clf.fit(X_train, y_train)
y_pred = linear_clf.predict(X_test)
score =accuracy_score(y_test,y_pred)
print("accuracy:   %0.3f" % score)
cm = confusion_matrix(y_test, y_pred)
cm
```
