```python
def get_results(df,ngram_range=(1,1),featuresize = 10000):
    previous_accuracy = 0.0
    for column in  df.drop('Label',axis = 1).columns:
    
        cv = CountVectorizer(max_features=featuresize,ngram_range=ngram_range)    
        x = cv.fit_transform(df[column].to_list()).toarray()
        y = df['Label'].values
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
    
        model =  MultinomialNB(alpha= 1)
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_pred,y_test)
        if (accuracy>previous_accuracy):
            print('column: {}'.format(column))
            print('featuresize: {}'.format(featuresize))
            print('ngram: {}'.format(ngram_range))
            print('accuracy: {}'.format(accuracy))
            print('')
            previous_accuracy = accuracy
    return previous_accuracy        
```

```python
get_results(df,ngram_range=(1,1),featuresize = 10000)
```

```python
features =[5000,8000,15000,20000,25000,30000,35000,40000,45000,50000,55000,60000]
grams = [(1,1),(1,2),(2,2),(1,3),(2,3),(3,3),(1,4),(2,4),(3,4),(4,4)]
for gram in grams:
    for feature in features:
        get_results(df,ngram_range=gram,featuresize = feature)
```
