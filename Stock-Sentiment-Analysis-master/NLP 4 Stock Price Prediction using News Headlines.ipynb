```python
#In case we would like to divide the data randomly

df = pd.read_csv('C:/Users/ishan/Desktop/Stock-Sentiment-Analysis-master/Data.csv', encoding = "ISO-8859-1")
df1 = df.copy()
```

```python
df.drop('Date',axis =1, inplace=True)
import random
x =df.index.to_list()
from sklearn.model_selection import train_test_split


train_indexes,test_indexes  = train_test_split(x, test_size = 0.15, random_state = 2)
```

```python
train = df.loc[train_indexes]
test  = df.loc[test_indexes]
print('Train Dataset:',train.shape)
print('Test Dataset:',test.shape)
```

```python
train.head()
```

```python
def get_x_y(data):                        #Text Preprocessing
    y_data = data['Label']
    x_data = data.drop('Label',axis = 1)
    x_data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
    columns = x_data.columns
    for column in columns:
        x_data[column]= x_data[column].str.lower()
    
    return x_data,y_data
x_train,y_train = get_x_y(train)
x_test,y_test = get_x_y(test)    
```

```python
# Converting all the headlines for a particular date into  single paragraphs

import re
import nltk

from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords 
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()


def get_headlines(data):
    headlines =[]
    paragraph =[]
    for i in range(0, len(data.index)):
        
        headlines.append(" ".join(str(x) for x in data.iloc[i,:]))
    for para in headlines:
        review = para.split()
        review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
        review = ' '.join(review)
        paragraph.append(review)
        
    return paragraph    
x_train_headlines =   get_headlines(x_train) 
x_test_headlines =   get_headlines(x_test) 
```

```python
from sklearn.feature_extraction.text import CountVectorizer   # For creating Bag of Words
from sklearn.feature_extraction.text import TfidfVectorizer  

from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
#countvector=CountVectorizer(ngram_range=(2,2))
countvector=TfidfVectorizer()

traindataset=countvector.fit_transform(x_train_headlines)
testdataset=countvector.transform(x_test_headlines)
```

```python

# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,y_train)
```

```python

y_pred = randomclassifier.predict(testdataset)
```

```python

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
score=accuracy_score(y_test,y_pred)
print(score)
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.naive_bayes import MultinomialNB
model =  MultinomialNB(alpha= 1)
model.fit(traindataset,y_train)
y_pred = model.predict(testdataset)
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
score=accuracy_score(y_test,y_pred)
print(score)
report=classification_report(y_test,y_pred)
print(report)
```

```python
y_pred = model.predict(traindataset)        # training dat accuracy
matrix=confusion_matrix(y_train,y_pred)
print(matrix)
```
